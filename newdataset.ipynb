```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('tw.csv')
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
df= data.drop(['user_desc','url','tweet_geo','tweet_location','tweet_entities','tweet_contributors'], axis='columns')
df.head(5)
```

```python
df.user_screen.describe()
```

```python
grouped= df.groupby('user_screen')
for x,y in grouped:
    print (x)
    print (y)
```

```python
newData= grouped[['user_screen','friends_count','tweet_followers_count','tweet_text']].agg({'friends_count': np.max, 'tweet_followers_count': np.max, 'tweet_text':  np.sum})
newData
```

```python
newData.to_csv('twdatabyuser.csv')
```
